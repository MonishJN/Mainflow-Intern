```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [150, 200, 250, 300, 350, 400],
    'Profit': [20, 30, 35, 50, 55, 70]
}
df=pd.DataFrame(data)
print(df)
```

```python
#Implementing and customizing Bar Graph
fig,ax1=plt.subplots()
ax1.set_xlabel('Month')
ax1.set_ylabel('Sales')
ax1.set_title("Bar Graph")
ax1.bar(df['Month'],df['Sales'],label='Sales')
ax1.legend(loc='upper left')
```

```python
#Implementing and customizing Linear Graph
fig,ax2=plt.subplots()
ax2.set_xlabel('Month')
ax2.set_ylabel('Profit')
ax2.set_title("Linear Graph")
ax2.plot(df['Month'],df['Profit'],label='Profit')
ax2.legend(loc="upper left")
```
