```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\yeraki vachuruken\USvideos.csv")
df.head(2)
```

```python
#Visualization and outliers

plt.figure()
sns.histplot(df["views"],bins=23)
plt.xlabel("Views")
plt.title("Histogram")
```

```python
plt.figure()
df["viewsbin"]=pd.cut(df["views"],bins=5)
sns.boxplot(x="viewsbin",y="likes",data=df)
```

```python
#Correaltions
plt.figure()
correaltion=df[["views","likes","dislikes"]].corr()
sns.heatmap(correaltion)
```

```python
sns.pairplot(df[["views","likes","dislikes"]])
```
